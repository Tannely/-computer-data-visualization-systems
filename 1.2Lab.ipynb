```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
data = pd.read_csv("shopping_trends.csv", encoding = 'cp1251')
data.head()
```

```python
"""1.Побудуйте гістограму віку клієнтів, використовуючи Seaborn"""

sns.histplot(data, x='Age', bins=10, kde=True, color='skyblue')
plt.title('Гістограма віку клієнтів')
plt.xlabel('Вік')
plt.ylabel('Кількість')
plt.show()
```

```python
"""2.Створіть стовпчасту діаграму або графік розподілу, щоб порівняти, які товари покупають клієнти найчастіше."""

items = data['Item Purchased'].value_counts().reset_index()
items.columns = ['Товар', 'Кількість']
#plt.figure(figsize=(8, 8))
sns.barplot(data=items.head(10), x = 'Кількість', y = 'Товар', palette='pastel')
plt.title('Найчастіше куповані товари')
plt.xlabel('Кількість')
plt.ylabel('Товар')
plt.tight_layout()  
plt.show()
```

```python
'''3.Використовуйте scatter plot, щоб візуалізувати залежність між віком та сумою покупок.'''

sns.scatterplot(data, x='Age', y='Purchase Amount (USD)', color='green')
plt.title('Залежність між віком та сумою покупок')
plt.xlabel('Вік')
plt.ylabel('Сума покупок')
plt.tight_layout()
plt.show()
```

```python
'''4. Графік розподілу оцінок товарів (review_rating): Побудуйте графік розподілу оцінок товарів для оцінки якості.'''

sns.histplot(data, x='Review Rating', bins=10, kde=True, color='skyblue')
plt.title('Розподіл оцінок товарів')
plt.xlabel('Оцінка')
plt.ylabel('Кількість')
plt.tight_layout()
plt.show()
```

```python
'''5. Створіть категоріальну діаграму, щоб відобразити розподіл покупок за статтею.'''

sns.countplot(data, x='Gender', palette='pastel')
plt.title('Розподіл покупок за статтю')
plt.xlabel('Стать')
plt.ylabel('Кількість покупок')
plt.tight_layout()
plt.show()
```

```python
'''6. Вивчіть, чи існує зв'язок між рейтингом товару і сумою покупок за допомогою scatter plot.'''

sns.scatterplot(data, x='Review Rating', y='Purchase Amount (USD)', alpha = 0.6, color='green')
plt.title('Зв\'язок між рейтингом товару та сумою покупок')
plt.xlabel('Рейтинг товару')
plt.ylabel('Сума покупок')
plt.tight_layout()
plt.show()
```

```python
'''7. Побудуйте стовпчату діаграму, яка відображає, в який сезон клієнти роблять найбільше покупок.'''

seasons = data['Season'].value_counts().reset_index()
seasons.columns =['Сезон', 'Кількість']
order_1 = ['Winter','Spring', 'Summer', 'Fall']
sns.barplot(data=seasons, x='Сезон', y='Кількість', order=order_1, palette='deep')

plt.title('Найбільш популярний сезон для покупок')
plt.xlabel('Сезон')
plt.ylabel('Кількість покупок')
plt.tight_layout()
plt.show()
```

```python
'''8. Створіть графік розподілу розмірів одягу, щоб визначити популярність різних розмірів.'''
sns.countplot(data, x = 'Size', palette='bright', order=data['Size'].value_counts().index)
plt.title('Розподіл розмірів одягу')
plt.xlabel('Розмір')
plt.ylabel('Кількість')

# Показати графік
plt.tight_layout()
plt.show()
```

```python
'''9. Порівняйте, скільки покупок робиться за різними методами оплати за допомогою стовпчатої діаграми'''

payment_methods = data['Payment Method'].value_counts().reset_index()
payment_methods.columns =['Метод оплати', 'Кількість']

sns.barplot(data = payment_methods, x='Метод оплати', y='Кількість', palette='pastel')
plt.title('Кількість покупок за різними методами оплати')
plt.xlabel('Метод оплати')
plt.ylabel('Кількість покупок')
plt.tight_layout()
plt.show()

```

```python
'''10. Залежність між кількістю попередніх покупок (previous_purchases) і сумою поточної покупки (purchase_amount):
Створіть scatter plot для визначення, чи існує зв'язок між кількістю 
попередніх покупок та сумою поточної покупки.'''

# Побудова scatter plot
sns.scatterplot(data, x='Previous Purchases', y='Purchase Amount (USD)', alpha = 0.6, color='pink')

# Налаштування заголовка та міток осей
plt.title('Залежність між кількістю попередніх покупок та сумою поточної покупки')
plt.xlabel('Кількість попередніх покупок')
plt.ylabel('Сума поточної покупки')

# Показати графік
plt.tight_layout()
plt.show()
```

Побудуйте графік поверхні (3D surface plot) для візуалізації залежності 
між трьома числовими змінними з використанням бібліотеки Seaborn.
Налаштуйте відображення графіка, включаючи назву графіку, підписи 
осей і інші необхідні параметри для кращого розуміння візуалізації.
Додайте кольорову легенду.

 


```python
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(x, y, z, cmap='viridis')
ax.set_title('3D Surface Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
fig.colorbar(surface, shrink=0.5, aspect=5)
plt.show()
```
